#### Gradient Descent with one variable

```python
import numpy as np  
import matplotlib.pyplot as plt  
```

Our function f(x) is as follows,

### f(x) = ùë•^4 ‚àí ùë•^3 ‚àí 6ùë•^2 + 2

```python
def derivativefunc(x):
    return eval('4*x**3-3*x**2-12*x')

def formula(x):
    return eval('x**4-x**3-6*x**2+2')
```

```python
def graph(formula, x_range):  
    x = np.array(x_range)  
    y = eval(formula)
    yValues.append(y)
    plt.plot(x, y)  
    plt.yticks(y)
    plt.grid(True, which='both')
    plt.show()
```

```python
def gradient(x,alpha,limit):
    xvals =[]
    func_vals=[]
    try:
        for _ in range(limit):
            deriv = derivativefunc(x)
            x = x-(alpha*deriv)
            xvals.append(x)
            func_vals.append(formula(x))  
        return [xvals,func_vals]
    except Exception as e:
        print e
        return [xvals,func_vals]
```

#### a. Graph this function for x in the interval [-3, 4].

```python
graph('x**4-x**3-6*x**2+2', range(-3, 5))
```

#### c. Gradient descent to this function

     x = x ‚Äì alpha * derivative(x),  where derivative is the derivative for our function f(x) 
     
 Setting x = -3 and alpha = 0.01, run gradient descent for 5 iterations 

```python
xvals, func_vals = gradient(-3,0.01,5)

plt.plot(xvals,func_vals)  
plt.grid(True, which='both')
plt.ylabel('f(x)')
plt.xlabel('x')
plt.show()
```

```python
print ('The 5 values of f(x) - {}. As you can see, it decreasing steadily'.format(func_vals))
```

#### d. Run the gradient descent again, starting with x = -3 for 1000 iterations

```python
xvals, func_vals = gradient(-3,0.01,1000)

result_01alpha = func_vals[:5] + func_vals[-5:]
x_01alpha = xvals[:5]+xvals[-5:]
plt.plot(x_01alpha,result_01alpha)  
plt.grid(True, which='both')
plt.ylabel('f(x)')
plt.xlabel('x')
plt.show()
```

```python
print ('The Last 5 values of f(x) - {}.'.format(func_vals[-5:]))
```

* The x value has converged to 1.4 
* As you can see, Gradient descent finds the local minimum but not the global minimum

#### e. Now with x=4 and alpha= 0.01

#### f. x=-3 and alpha=0.05

```python
xvals, func_vals = gradient(-3,0.05,1000)

result_05alpha = func_vals[:5] + func_vals[-5:]
x_05alpha = xvals[:5]+xvals[-5:]

plt.plot(x_05alpha,result_05alpha)  
plt.grid(True, which='both')
plt.ylabel('f(x)')
plt.xlabel('x')
plt.show()
```

##### comparing two results
*  x=-3 and alpha=0.01 
*  x=-3 and alpha=0.05

```python
plt.plot(range(10), result_01alpha)
plt.plot(range(10),result_05alpha)
plt.grid(True, which='both')
plt.legend(['alpha = 0.01', 'alpha = 0.05'], loc='upper right')
plt.ylabel('f(x)')
plt.xlabel('# of Iterations - (1st five and last five values)')
plt.show()
```

Thus, we can see that setting alpha=0.01 doesn't find the global minimum, but on setting the learning rate higher as 0.05, helps to find the global minimum of f(x).
